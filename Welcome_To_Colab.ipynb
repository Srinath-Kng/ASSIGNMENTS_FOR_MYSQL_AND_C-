```
!pip install -U transformers
from transformers import pipeline

# Use text2text-generation for question generation tasks
nlp = pipeline("text2text-generation", model="valhalla/t5-base-qg-hl")
nlp2 = pipeline("text2text-generation", model="valhalla/t5-base-qa-qg-hl")
nlp3 = pipeline("text2text-generation", model="valhalla/t5-base-e2e-qg")

```

```
text1 = """
Once upon a time, in a land far away, there was a magician who could create the most wonderful illusions.
People from all over the world would come to see his magic shows.
"""
questions = nlp(text1)
for i in range(len(questions)):
    print("Question:", questions[i]['generated_text'])
```

```
qa_results = nlp2(text1)
for result in qa_results:
    print("Generated Text:", result['generated_text'])
```

```
e2e_results = nlp3(text1)
for result in e2e_results:
    print("Generated Text:", result['generated_text'])
```

```
from transformers import pipeline

# Initialize the question-answering pipeline
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# Define your text
text1 = """
Once upon a time, in a land far away, there was a magician who could create the most wonderful illusions.
People from all over the world would come to see his magic shows.
"""

# Define your question
question = "What does a magician create?"

# Use the pipeline for question answering
qa_result = qa_pipeline({
    "question": question,
    "context": text1
})

# Print the answer
print("Answer:", qa_result['answer'])

```
