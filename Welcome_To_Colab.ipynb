```
from transformers import pipeline

# Initialize the question generation pipeline
question_generator = pipeline('text2text-generation', model='t5-small')

# Initialize the question answering pipeline
qa_pipeline = pipeline('question-answering')

# Sample text input
text = """
Artificial intelligence (AI) is intelligence demonstrated by machines, in contrast to the natural intelligence displayed by humans and animals. Leading AI textbooks define the field as the study of "intelligent agents": any device that perceives its environment and takes actions that maximize its chance of successfully achieving its goals.
"""

# Define a function to generate questions
def generate_questions(text):
    questions = question_generator(text, max_new_tokens=50, truncation=True)
    return questions

# Define a function to generate answers
def generate_answers(questions, context):
    answers = []
    for question in questions:
        question_text = question['generated_text']
        answer = qa_pipeline(question=question_text, context=context)
        answers.append((question_text, answer['answer']))
    return answers

# Generate questions
questions = generate_questions(text)

# Generate answers for each question
qa_pairs = generate_answers(questions, text)

# Print questions and answers
for question, answer in qa_pairs:
    print(f"Question: {question}")
    print(f"Answer: {answer}\n")
```

```
from transformers import pipeline
import random

# Initialize the pipelines
question_generator = pipeline('text2text-generation', model='t5-small')
qa_pipeline = pipeline('question-answering')

# Sample text input
text = """
Artificial intelligence (AI) is intelligence demonstrated by machines, in contrast to the natural intelligence displayed by humans and animals. Leading AI textbooks define the field as the study of "intelligent agents": any device that perceives its environment and takes actions that maximize its chance of successfully achieving its goals.
"""

# Define a function to generate questions
def generate_questions(text):
    questions = question_generator(text, max_new_tokens=50, truncation=True)
    return questions

# Define a function to generate multiple-choice options
def generate_options(question, context):
    # Use the QA model to find the correct answer
    correct_answer = qa_pipeline(question=question, context=context)['answer']

    # Generate distractors using semantic similarity or other techniques
    distractors = [
        "natural agents",
        "artificial devices",
        "intelligent machines"
    ]

    # Randomly select distractors
    options = random.sample(distractors, 2)
    options.append(correct_answer)
    random.shuffle(options)

    return options, correct_answer

# Generate questions
questions = generate_questions(text)

# Generate multiple-choice questions with options
for question in questions:
    question_text = question['generated_text']
    options, correct_answer = generate_options(question_text, text)

    print(f"Question: {question_text}")
    for i, option in enumerate(options, 1):
        print(f"Option {i}: {option}")
    print(f"Correct Answer: {correct_answer}\n")
```

```
from transformers import pipeline
import random

# Initialize the pipeline for text summarization or key phrase extraction
summarizer = pipeline('summarization', model='facebook/bart-large-cnn')

# Sample text input
text = """
Artificial intelligence (AI) is intelligence demonstrated by machines, in contrast to the natural intelligence displayed by humans and animals. Leading AI textbooks define the field as the study of "intelligent agents": any device that perceives its environment and takes actions that maximize its chance of successfully achieving its goals.
"""

# Function to generate fill-in-the-blank questions
def generate_fill_in_the_blank(text):
    # Summarize or extract key phrases
    summary = summarizer(text, max_length=50, min_length=25, do_sample=False)[0]['summary_text']

    # Split the summary into words
    words = summary.split()

    # Randomly select a word to replace with a blank
    if len(words) > 5:  # Ensure there's enough context
        blank_index = random.randint(0, len(words) - 1)
        answer = words[blank_index]
        words[blank_index] = "______"
        question = " ".join(words)
    else:
        question = summary
        answer = "N/A"

    return question, answer

# Generate fill-in-the-blank question
question, answer = generate_fill_in_the_blank(text)

# Print the fill-in-the-blank question and answer
print(f"Fill-in-the-Blank Question: {question}")
print(f"Answer: {answer}")
```

```
from transformers import pipeline
import random

# Sample text input
text = """
Artificial intelligence (AI) is intelligence demonstrated by machines, in contrast to the natural intelligence displayed by humans and animals. Leading AI textbooks define the field as the study of "intelligent agents": any device that perceives its environment and takes actions that maximize its chance of successfully achieving its goals.
"""

# Function to generate true or false questions
def generate_true_false_questions(text):
    # Split the text into sentences
    sentences = text.strip().split('.')

    # Create a list to store true/false questions
    true_false_questions = []

    # Iterate over each sentence
    for sentence in sentences:
        sentence = sentence.strip()
        if sentence:
            # Randomly decide to alter the statement to create a false version
            is_true = random.choice([True, False])
            if not is_true:
                # Alter the statement to make it false
                altered_sentence = sentence.replace("machines", "humans")
                true_false_questions.append((altered_sentence, False))
            else:
                true_false_questions.append((sentence, True))

    return true_false_questions

# Generate true or false questions
questions = generate_true_false_questions(text)

# Print the true or false questions and answers
for i, (question, is_true) in enumerate(questions, 1):
    print(f"Question {i}: {question}?")
    print(f"Answer: {'True' if is_true else 'False'}\n")
```
